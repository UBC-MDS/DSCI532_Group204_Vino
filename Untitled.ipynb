```python
import altair as alt
from vega_datasets import data
import pandas as pd
```

```python
states = alt.topo_feature(data.us_10m.url, 'states')
source = data.population_engineers_hurricanes.url

alt.Chart(states).mark_geoshape().encode(
    alt.Color('engineers:Q', type='quantitative',
             scale=alt.Scale(range=['purple', 'red']),
             legend=alt.Legend(title="Wine Quality"))
).transform_lookup(
    lookup='id',
    from_=alt.LookupData(source, 'id', ['engineers'])
).project(
    type='albersUsa'
).properties(
    width=500,
    height=300
).resolve_scale(
    color='independent'
)
```

```python
source = pd.DataFrame({
    'Winery': ['1', '2', '3', '4', '5'],
    'Average Rating': [98, 94, 78, 65, 22]
})

alt.Chart(source).mark_bar(color="red", opacity=0.7).encode(
    alt.X('Winery'),
    alt.Y('Average Rating',
        sort=alt.EncodingSortField(
            field='Average Rating',
            #op="sum",
            order='descending')),
    alt.Color('Winery', scale=alt.Scale(range=['red', 'grey',' grey', 'grey', 'grey']), legend=None)
).properties(height=300, width=450).configure_axis(
    grid=False,
    labelFontSize=12,
    titleFontSize=20,
    labelFontStyle='Arial'
)
```
